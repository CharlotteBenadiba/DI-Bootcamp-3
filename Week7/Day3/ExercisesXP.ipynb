##  Exercise 1: Implementing And Comparing Normalization Techniques
Instructions
Create a synthetic dataset with two features: FeatureA and FeatureB, where FeatureA is in the range of [0, 1000] and FeatureB is in the range of [0, 10].
Apply Min-Max Scaling to this dataset and visualize the results using a scatter plot.
Apply Z-score Normalization to the same dataset and visualize these results as well.
Write a short analysis comparing the effects of both normalization techniques on the dataset. Discuss which technique might be more appropriate based on the distribution and range of the data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from scipy import stats
from scipy.stats import zscore
from scipy.stats.mstats import winsorize
```

```python
# Create a synthetic dataset
np.random.seed(42)
num_samples = 100
feature_a = np.random.uniform(0, 1000, num_samples)
feature_b = np.random.uniform(0, 10, num_samples)

data = pd.DataFrame({'FeatureA': feature_a, 'FeatureB': feature_b})

# Apply Min-Max Scaling
scaler_minmax = MinMaxScaler()
data_minmax = scaler_minmax.fit_transform(data)
data_minmax = pd.DataFrame(data_minmax, columns=['FeatureA', 'FeatureB'])

# Apply Z-score Normalization
scaler_zscore = StandardScaler()
data_zscore = scaler_zscore.fit_transform(data)
data_zscore = pd.DataFrame(data_zscore, columns=['FeatureA', 'FeatureB'])
```

```python
# Visualize the results using scatter plots
plt.figure(figsize=(12, 5))

plt.subplot(1, 3, 1)
plt.scatter(data['FeatureA'], data['FeatureB'])
plt.title('Original data')
plt.xlabel('FeatureA')
plt.ylabel('FeatureB')

plt.subplot(1, 3, 2)
plt.scatter(data_minmax['FeatureA'], data_minmax['FeatureB'])
plt.title('Min-Max scaling')
plt.xlabel('FeatureA (Scaled)')
plt.ylabel('FeatureB (Scaled)')

plt.subplot(1, 3, 3)
plt.scatter(data_zscore['FeatureA'], data_zscore['FeatureB'])
plt.title('Z-score normalization')
plt.xlabel('FeatureA (Normalized)')
plt.ylabel('FeatureB (Normalized)')

plt.tight_layout()
plt.show()
```

<div style="border:solid green 2px; padding: 20px">

- Min-Max Scaling: this technique scales the features to a specific range (typically [0, 1]). In the scatter plot, we can see that the points are now confined within the [0, 1] range for both features.
- Z-score Normalization: this technique standardizes the features to have a mean of 0 and a standard deviation of 1. In the scatter plot, we can observe that the points are centered around 0 and have a standard deviation of 1 for both features.
- The distribution in all three graphs looks the same. If the distribution and range of the features are important, and we want to maintain the original range, Min-Max Scaling might be more appropriate. If we want to standardize the features to have a mean of 0 and a standard deviation of 1, especially if the features have different units or magnitudes, Z-score Normalization might be preferred.

##  Exercise 2 : Analyzing The Impact Of Min-Max Scaling On Synthetic Data
Instructions
Create a synthetic dataset with two features: FeatureA and FeatureB. Let FeatureA range from 1 to 100 and FeatureB range from 1,000 to 10,000.
Apply Min-Max Scaling to this dataset using Python and the MinMaxScaler from Scikit-Learn.
Visualize the original and scaled datasets using line plots or scatter plots (using Matplotlib or Seaborn).
Write a brief analysis on how Min-Max Scaling transformed the data, focusing on the change in the range of values for both features.

```python
# Create a synthetic dataset
np.random.seed(42)
num_samples1 = 100
feature_a1 = np.random.uniform(1, 100, num_samples1)
feature_b1 = np.random.uniform(1000, 10000, num_samples1)

data1 = pd.DataFrame({'FeatureA': feature_a1, 'FeatureB': feature_b1})

# Apply Min-Max Scaling
scaler_minmax1 = MinMaxScaler()
data_scaled = scaler_minmax1.fit_transform(data1)
data_scaled = pd.DataFrame(data_scaled, columns=['FeatureA', 'FeatureB'])
```

```python
# Visualize the results using line plots
plt.figure(figsize=(11, 5))

plt.subplot(1, 2, 1)
plt.plot(data1['FeatureA'], label='FeatureA')
plt.plot(data1['FeatureB'], label='FeatureB')
plt.title('Original data')
plt.xlabel('Sample index')
plt.ylabel('Values')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(data_scaled['FeatureA'], label='FeatureA (Scaled)')
plt.plot(data_scaled['FeatureB'], label='FeatureB (Scaled)')
plt.title('Min-Max scaling')
plt.xlabel('Sample index')
plt.ylabel('Scaled values')
plt.legend()

plt.tight_layout()
plt.show()
```

<div style="border:solid green 2px; padding: 20px">
    
- After applying Min-Max Scaling, both FeatureA and FeatureB are transformed to a common scale - [0, 1].
The transformation ensures that the values of both features are proportionally adjusted to the range [0, 1].
- Min-Max Scaling compresses the original range of each feature to the specified range ([0, 1]). The relative relationships between data points within each feature are preserved, but the absolute values are transformed.
Min-Max Scaling effectively transformed the data, bringing both features to a common scale, which can be useful in scenarios where the magnitude or range of features needs to be standardized for certain machine learning algorithms or analyses.
- In the first graph, the two features have different scales, and the feature with smaller values is uninformative. In the second graph, both features are in the same range and they can be compared. 

## Exercise 3: Experimenting With Z-Score Normalization On Simulated Data
Instructions
Generate a simulated dataset with two features: FeatureX and FeatureY. Let FeatureX be normally distributed around a mean of 50 with a standard deviation of 10, and FeatureY be normally distributed around a mean of 500 with a standard deviation of 100. Apply Z-score Normalization to this dataset using Python and the StandardScaler from Scikit-Learn.
Create visualizations (such as histograms or box plots) of the original and normalized datasets to compare their distributions.
Discuss the effects of Z-score Normalization on the dataset, particularly how it alters the mean and standard deviation of the features.

```python
# Generate a simulated dataset
np.random.seed(42)
num_samples2 = 1000
feature_x = np.random.normal(loc=50, scale=10, size=num_samples2)
feature_y = np.random.normal(loc=500, scale=100, size=num_samples2)

data2 = pd.DataFrame({'FeatureX': feature_x, 'FeatureY': feature_y})

# Apply Z-score Normalization
scaler_zscore = StandardScaler()
data_normalized = scaler_zscore.fit_transform(data2)
data_normalized = pd.DataFrame(data_normalized, columns=['FeatureX', 'FeatureY'])
```

```python
# Visualize the results using histograms
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(data2['FeatureX'], kde=True, color='blue')
sns.histplot(data2['FeatureY'], kde=True, color='orange')
plt.title('Original data distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.histplot(data_normalized['FeatureX'], kde=True, color='blue')
sns.histplot(data_normalized['FeatureY'], kde=True, color='orange')
plt.title('Z-Score normalized distribution')
plt.xlabel('Standardized values')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

<div style="border:solid green 2px; padding: 20px">
    
- The histograms show the original distributions of FeatureX and FeatureY.
- After applying Z-score Normalization, both FeatureX and FeatureY are transformed to have a mean of 0 and a standard deviation of 1. The histograms show the standardized distributions of FeatureX and FeatureY.
- The mean of each feature is shifted to 0, and the standard deviation becomes 1. This is a standardization process.
Z-score Normalization does not change the shape of the original distributions but places them on a common scale, making it easier to compare and analyze features with different units and scales.
Z-score Normalization standardizes the features, bringing them to a common scale, and is particularly useful when features have different magnitudes or units.

## Exercise 4 : Identifying Outliers With Box Plots And Z-Scores
Instructions
Dataset Suggestion: Use the Boston Housing Dataset available on Kaggle. This dataset contains information about housing in the area of Boston Mass. Its a good dataset for practicing outlier detection as it has a variety of features with different scales and distributions.

Load the Boston Housing Dataset into a Python environment.
Use box plots to visually identify potential outliers in various features like CRIM (crime rate), RM (average number of rooms per dwelling), and MEDV (Median value of owner-occupied homes in $1000s).
Calculate Z-scores for these features and identify data points that have a Z-score higher than 3 or lower than -3.
Reflect on how outliers identified by box plots and Z-scores overlap or differ.

```python
boston_housing = pd.read_csv('HousingData.csv')

boston_housing.info()
boston_housing.head(10)
```

```python
# Visualize potential outliers using box plots
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.boxplot(y=boston_housing['CRIM'])
plt.title('Box plot - CRIM')

plt.subplot(1, 3, 2)
sns.boxplot(y=boston_housing['RM'])
plt.title('Box plot - RM')

plt.subplot(1, 3, 3)
sns.boxplot(y=boston_housing['MEDV'])
plt.title('Box plot - MEDV')

plt.tight_layout()
plt.show()
```

```python
# Calculate Z-scores for specified features
scaler = StandardScaler()
boston_housing_zscore_normalized = pd.DataFrame(scaler.fit_transform(boston_housing[['CRIM', 'RM', 'MEDV']]), columns=['CRIM', 'RM', 'MEDV'])

# Identify data points with Z-scores higher than 3 or lower than -3
outliers = boston_housing_zscore_normalized[(boston_housing_zscore_normalized > 3) | (boston_housing_zscore_normalized < -3)].dropna()

print("Identified outliers:")
print(outliers)
```

```python
# another way to check outliers
features = ['CRIM', 'RM', 'MEDV']
df_zscores = boston_housing[features].apply(zscore)

# Identify data points with Z-scores higher than 3 or lower than -3
outliers_boston_housing = df_zscores[(df_zscores > 3) | (df_zscores < -3)].dropna()

print("Identified outliers:")
outliers_boston_housing
```

<div style="border:solid green 2px; padding: 20px">
    
- Box plots provide a visual representation of the distribution of data and help identify potential outliers. The CRIM column has the most emissions. The "box" itself is in the range from 0 to 5, and the emissions go up to 90. In the RM and MEDV columns strong outliers are also noticeable.
- Z-scores quantify how many standard deviations a data point is from the mean. Z-scores above 3 or below -3 are often considered as potential outliers. However, the calculation of Z-scores did not reveal emissions greater than 3 or less than -3.
- Differences may occur due to the sensitivity of each method to the characteristics of the data.

##  Exercise 5 : Handling Outliers In A Large Dataset
Instructions
Dataset Suggestion: New York City Airbnb Open Data on Kaggle. This dataset provides a comprehensive look at the Airbnb activity in New York City and includes features like price, number of reviews, availability, etc., which are ideal for practicing outlier handling.

Load the New York City Airbnb dataset.
Identify outliers in the price column using the IQR method.
Apply three different outlier handling techniques: trimming, capping, and transformation.
Analyze the effect of each technique on the distribution of the price feature.
Discuss the pros and cons of each technique in the context of this dataset.

```python
ny_airbnb = pd.read_csv('AB_NYC_2019.csv')

ny_airbnb.info()
ny_airbnb.head(10)
```

```python
# Calculate the Interquartile Range (IQR)
Q1 = ny_airbnb['price'].quantile(0.25)
Q3 = ny_airbnb['price'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers_ny_airbnb = ny_airbnb[(ny_airbnb['price'] < lower_bound) | (ny_airbnb['price'] > upper_bound)]

print("Identified outliers:")
outliers_ny_airbnb
```

<div style="border:solid green 2px; padding: 20px">
The outliers in the price' column using the IQR method are 2972 rows (out of 48,895), which is about 6% of the total dataset.

```python
# Method 1: Trimming
trimmed_df = ny_airbnb[(ny_airbnb['price'] >= lower_bound) & (ny_airbnb['price'] <= upper_bound)]
trimmed_df
```

```python
# Method 2: Capping
winsorized_df = ny_airbnb.copy()
winsorized_df['price'] = winsorize(ny_airbnb['price'], limits=(0.05, 0.05))
winsorized_df
```

```python
# Method 3: Transformation
transformed_df = ny_airbnb.copy()
transformed_df['price'], _ = stats.boxcox(ny_airbnb['price'] + 1)
transformed_df
```

```python
# Visualize the distribution of 'price' before handling outliers and after using methods
plt.figure(figsize=(12, 6))
plt.subplot(1, 4, 1)
sns.boxplot(x=ny_airbnb['price'])
plt.title('Before handling outliers')

plt.subplot(1, 4, 2)
sns.boxplot(x=trimmed_df['price'])
plt.title('After trimming')

plt.subplot(1, 4, 3)
sns.boxplot(x=winsorized_df['price'])
plt.title('After Winsorizing')

plt.subplot(1, 4, 4)
sns.boxplot(x=transformed_df['price'])
plt.title('After Box-Cox transformation')

plt.tight_layout()
plt.show()
```

<div style="border:solid green 2px; padding: 20px">
    
- Outliers were removed the most when using the IQR method - the dataset includes 45923 rows, and after using the other two methods, the dataset includes 48895 rows.
- The most informative boxplots after applying trimming and Winsorizing.
- Trimming:
Pros:
Simple and straightforward.
Preserves the original data without transforming it.
Cons:
Discards data points, potentially leading to information loss.
May not be suitable if the dataset is already small.
    
Capping (Winsorizing):
Pros:
Retains all data points but compresses extreme values.
Preserves the overall structure of the dataset.
Cons:
The choice of the capping limits is somewhat arbitrary and may impact results.
May not handle extremely skewed data effectively.
    
Transformation (Box-Cox):
Pros:
Allows for a variety of transformations to stabilize variance and make the data more normal.
Suitable for handling right-skewed distributions.
Cons:
Assumes that the data contains only positive values.
May not perform well if the data distribution is not well-approximated by a power-law distribution.
The Box-Cox transformation involves finding an optimal lambda value, which might not always be straightforward.

## Exercise 6 : Impact Of Outliers On Data Analysis
Instructions
Dataset Suggestion: Wine Quality Dataset on Kaggle. This dataset contains physicochemical tests of wines and a quality rating, making it suitable for understanding the impact of outliers.

Load the Wine Quality Dataset.
Identify outliers in key features like fixed acidity, residual sugar, and alcohol using your preferred method.
Calculate the mean and standard deviation of these features with and without outliers.
Analyze how outliers affect the interpretation of these features.
Reflect on the importance of considering outliers in data analysis, especially in a context like wine quality assessment.

```python
wine_qt = pd.read_csv('WineQT.csv')

wine_qt.info()
wine_qt.head(10)
```

```python
# General information about three features with outliers
print(wine_qt['fixed acidity'].describe())
print('------')
print(wine_qt['residual sugar'].describe())
print('------')
print(wine_qt['alcohol'].describe())
```

```python
# Identify outliers using Z-score
z_scores1 = zscore(wine_qt[['fixed acidity', 'residual sugar', 'alcohol']])
outliers_mask = (z_scores1 > 3) | (z_scores1 < -3)

# Calculate mean and standard deviation with and without outliers
mean_with_outliers = wine_qt[['fixed acidity', 'residual sugar', 'alcohol']].mean()
std_with_outliers = wine_qt[['fixed acidity', 'residual sugar', 'alcohol']].std()

# Remove outliers
df_no_outliers = wine_qt[~outliers_mask.any(axis=1)]
display(df_no_outliers)

# Calculate mean and standard deviation without outliers
mean_no_outliers = df_no_outliers[['fixed acidity', 'residual sugar', 'alcohol']].mean()
std_no_outliers = df_no_outliers[['fixed acidity', 'residual sugar', 'alcohol']].std()

print("Mean and standard deviation with outliers:")
print(mean_with_outliers)
print(std_with_outliers)

print("\nMean and standard deviation without outliers:")
print(mean_no_outliers)
print(std_no_outliers)
```

<div style="border:solid green 2px; padding: 20px">
    
- There are 37 outliers in the dataset, which is about 3.2% of the entire dataset.The mean and standard deviation decreased in each of the three columns, but not significantly.
- With outliers included, the mean can be heavily influenced by extreme values. The standard deviation, which measures the spread of the data, is also affected.
In the presence of outliers, the mean tends to be higher, and the standard deviation increases, indicating greater variability in the data.
The presence of outliers can distort the normal distribution of the data, making it skewed. This affects the accuracy of statistical summaries and visualizations.
Without proper handling, outliers can lead to misleading interpretations about the central tendency and spread of the data.
Removing outliers (trimming or transforming) can help obtain a more representative picture of the typical values within the dataset.
Outlier removal is particularly crucial in tasks like wine quality assessment, where extreme values might be indicative of errors or rare conditions rather than typical characteristics.
- Wine quality assessment involves identifying key factors that contribute to the overall quality of the wine.
Outliers in features like acidity, sugar content, or alcohol levels might indicate rare or exceptional cases that are not representative of the general wine population.
Ignoring outliers could lead to biased assessments and inaccurate models, impacting the reliability of quality predictions.
Outliers can significantly impact the performance of statistical models. Robust models that are less sensitive to outliers may be preferred in such scenarios.
Understanding the nature of outliers helps in making informed decisions about whether to retain, transform, or remove them based on the specific objectives of the analysis.
